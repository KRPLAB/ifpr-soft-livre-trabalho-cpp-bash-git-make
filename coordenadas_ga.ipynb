```python
import numpy as np

def car_ci(x, y, z):
    r = np.sqrt(x**2 + y**2)
    te = np.arctan2(y, x)
    return r, te, z

def car_esf(x, y, z):
    ro = np.sqrt(x**2 + y**2 + z**2)
    te = np.arctan2(y, x)
    phi = np.arctan2(np.sqrt(x**2 + y**2), z)
    return ro, te, phi

def cil_car(r, te, z):
    x = r * np.cos(te)
    y = r * np.sin(te)
    return x, y, z

def cil_esf(r, te, z):
    ro = np.sqrt(r**2 + z**2)
    te_e = te
    phi = np.arctan2(r, z)
    return ro, te_e, phi

def esf_car(ro, te, phi):
    x = ro * np.sin(phi) * np.cos(te)
    y = ro * np.sin(phi) * np.sin(te)
    z = ro * np.cos(phi)
    return x, y, z

def esf_cil(ro, te_e, phi):
    r = ro * np.sin(phi)
    te_c = te_e
    z = ro * np.cos(phi)
    return r, te_c, z

def run_converter():
    while True:
        print("\n--- Conversor de Coordenadas ---")
        print("Converter de:")
        print("1 - Cartesianas (x, y, z)")
        print("2 - Cilíndricas (r, theta, z)")
        print("3 - Esféricas (ro, theta, phi)")
        print("0 - Sair")

        origem_opcao = input("Origem: ").upper()
        if origem_opcao == '0':
            break

        if origem_opcao not in ['1', '2', '3']:
            print("Erro: Opção de origem inválida.")
            continue

        print("Converter para:")
        print("A - Cartesianas")
        print("B - Cilíndricas")
        print("C - Esféricas")

        destino_opcao = input("Destino: ").upper()

        if destino_opcao not in ['A', 'B', 'C']:
            print("Erro: Opção de destino inválida.")
            continue

        print("-" * 25)

        try:
            # De Cartesianas para...
            if origem_opcao == '1':
                x = float(input("X: "))
                y = float(input("Y: "))
                z = float(input("Z: "))

                if destino_opcao == 'B':  # Cartesianas para Cilíndricas
                    r, te, z_cil = car_ci(x, y, z)
                    print(f"\nCilíndricas: (r={r:.2f}, theta={np.degrees(te):.2f}°, z={z_cil:.2f})")
                elif destino_opcao == 'C':  # Cartesianas para Esféricas
                    ro, te, phi = car_esf(x, y, z)
                    print(f"\nEsféricas: (ro={ro:.2f}, theta={np.degrees(te):.2f}°, phi={np.degrees(phi):.2f}°)")

            # De Cilíndricas para...
            elif origem_opcao == '2':
                r = float(input("R: "))
                te = np.radians(float(input("Theta (graus): ")))
                z = float(input("Z: "))

                if destino_opcao == 'A':  # Cilíndricas para Cartesianas
                    x, y, z_cart = cil_car(r, te, z)
                    print(f"\nCartesianas: (x={x:.2f}, y={y:.2f}, z={z_cart:.2f})")
                elif destino_opcao == 'C':  # Cilíndricas para Esféricas
                    ro, te_e, phi = cil_esf(r, te, z)
                    print(f"\nEsféricas: (ro={ro:.2f}, theta={np.degrees(te_e):.2f}°, phi={np.degrees(phi):.2f}°)")

            # De Esféricas para...
            elif origem_opcao == '3':
                ro = float(input("Ro: "))
                te = np.radians(float(input("Theta (graus): ")))
                phi = np.radians(float(input("Phi (graus): ")))

                if destino_opcao == 'A':  # Esféricas para Cartesianas
                    x, y, z_cart = esf_car(ro, te, phi)
                    print(f"\nCartesianas: (x={x:.2f}, y={y:.2f}, z={z_cart:.2f})")
                elif destino_opcao == 'B':  # Esféricas para Cilíndricas
                    r, te_c, z_cil = esf_cil(ro, te, phi)
                    print(f"\nCilíndricas: (r={r:.2f}, theta={np.degrees(te_c):.2f}°, z={z_cil:.2f})")

        except ValueError:
            print("Erro: Digite apenas números.")

if __name__ == "__main__":
    run_converter()
```
